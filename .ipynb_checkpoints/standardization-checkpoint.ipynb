```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read file 
df_norm = pd.read_csv('alex_normal_data.csv', index_col=0)
df_norm.head()
```

```python
df_total = pd.read_csv('alex_total_df.csv', index_col=0)
df_total.head()
```

```python
df_ids = df_total.filter(['I1', 'I2', 'I3'])
df_ids.head()
```

```python
df_newcols = df_total.filter(['P(mid)', 'C3p', 'C5p', 'C6p', 'LongSentences', 'RealWords', 'LongWords', 'PositiveWords', 'NegativeWords', 'UncertainWords'])
df_newcols.head()
```

```python
df_new_norm = pd.concat([df_norm, df_newcols], axis=1)
df_new_norm.head()
```

```python
# do descriptive stats to confirm correct dataset has been loaded
# mins = 0 and maxs = 1 
# mean and stdev is not where we want it to be 
df_new_norm.describe()
```

```python
# show distribution of one column
# notice the scale on the x-axis 
# varible is still skewed 
plt.hist(df_new_norm['P(IPO)'], range=(0,1))
plt.show()
```

```python
df_norm.skew()
```

```python
from sklearn import preprocessing
# define column names
names = df_new_norm.columns

# create scaler variable 
scaler = preprocessing.StandardScaler()

# fit and scale data 
df_standard = scaler.fit_transform(df_new_norm)
df_standard = pd.DataFrame(df_standard, columns=names)
```

```python
# check if data was scaled correctly
# mean and stdev should be equal to or almost equal to 0 and 1 respectively 
df_standard.describe()
```

```python
# notice different scale on the x-axis
# data has bee scaled correctly 
plt.hist(df_standard['P(IPO)']), #range=(-2,2))
plt.show()
```

```python
# export standardized data to a csv 
df_standard.to_csv('standard_data.csv')
```

```python
# examine distribution for all variables 
for col in df_standard.columns:
    df_standard.hist(color='blue', range=(-2,2))
    break
plt.show()
```
