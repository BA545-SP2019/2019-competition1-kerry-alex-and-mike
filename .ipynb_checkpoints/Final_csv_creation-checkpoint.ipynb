```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('csv_files/p3feature-selection.csv',index_col=0)
df.head()
```

```python
df = df.drop(['C1','C2','C3p','C7','C5p','T4p'],axis=1)
df.head()
```

```python
df2 = df.filter(['S2p'])
df2.head()
```

```python
df = df.drop('S2p',axis=1)
df.head()
```

```python
df = pd.concat([df2, df],axis=1)
df.head()
```

```python
df.to_csv('final.csv')
```
