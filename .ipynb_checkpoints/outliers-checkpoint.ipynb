```python
upper = []
lower = []

for i in df_values.columns:
    Q1 = df_values[i].quantile(0.25)
    Q3 = df_values[i].quantile(0.75)
    IQR = Q3 - Q1
    upper_list = Q3 + 1.5 * IQR
    upper.append(upper_list)
    lower_list = Q1 - 1.5 * IQR
    lower.append(lower_list)
    

#for i in range(len(df_values.columns)):
#    for x in df_values[df_values.columns[i]]:
#        if x > upper[i]:
#            x = upper[i]
#        elif x < lower[i]:
#            x = lower[i]
```

```python
def outlier_handling(x, upper, lower):
    if x > upper:
        x = upper
    elif x < lower:
            x = lower
    else:
        pass
    return x
```

```python
for col in df_values.columns:
    q1 = df_values.loc[:, col].quantile(0.25)
    q3 = df_values.loc[:, col].quantile(0.75)
    IQR = q3 - q1
    lbound = q1 - 1.5*IQR
    ubound = q3 + 1.5*IQR
    df_values.loc[:, col] = df_values.loc[:, col].apply(lambda x: outlier_handling(x, ubound,lbound))
```
