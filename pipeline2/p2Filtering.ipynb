```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
# load the dataset
df = pd.read_csv('../csv_files/p2feature-selection.csv', index_col=0)
df.head()
```

```python
#Converting the datagrame into the correct arrays to make filtering easier 
array = df.values
X = abs(array[:,:-2]) #features
Y1 = array[:,-2]
Y2 = array[:,-1]
```

```python
# Feature selection for Target 1
test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, Y1)

# Summarize scores
np.set_printoptions(precision=3)
print(fit.scores_)

features = fit.transform(X)
# Summarize selected features
print(features[0:5,:])
```

```python
# Feature selection for Target 2
test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, Y2)

# Summarize scores
np.set_printoptions(precision=3)
print(fit.scores_)

features = fit.transform(X)
# Summarize selected features
print(features[0:5,:])
```
