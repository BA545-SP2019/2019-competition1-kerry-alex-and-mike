```python
# import packages 
import numpy as np
import pandas as pd
```

```python
# read in csv
df = pd.read_csv('../csv_files/p2binning.csv', index_col=0)
df.head()
```

```python
# verify that the columns are named accurately 
df.columns
```

```python
# filter out the x variables 
dfx = df.filter(['C1', 'C4', 'C6p', 'C7','C5p', 'T4p', 'T3p', 'T5p', 'S1p', 'S2p', 'S3p', 'C2', 'C3p', 'Industry_Bins'])
dfx.head()
```

```python
# filter out the target variables 
dfy = df.filter(['Y1', 'Y2'])
dfy.head()
```

```python
# concat the dataframes 
# this is the dataframe we will use in feature selection
df = pd.concat([dfx, dfy], axis=1)
df.head()
```

```python
df.to_csv('../csv_files/p2feature-selection.csv')
```
