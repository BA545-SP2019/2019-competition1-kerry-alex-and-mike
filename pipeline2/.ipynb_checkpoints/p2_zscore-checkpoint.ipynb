```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read file 
dfx = pd.read_csv('csv_files/x.csv', index_col=0)
dfx.head()
```

```python
# do descriptive stats to confirm correct dataset has been loaded
# mins = 0 and maxs = 1 
# mean and stdev is not where we want it to be 
dfx.describe()
```

```python
dfx.skew()
```

```python
# show distribution of one column
# notice the scale on the x-axis 
# varible is still skewed 
plt.hist(dfx['C1'])
plt.show()
```

```python
from sklearn import preprocessing
# define column names
names = dfx.columns

# create scaler variable 
scaler = preprocessing.StandardScaler()

# fit and scale data 
dfxstandard = scaler.fit_transform(dfx)
dfxstandard = pd.DataFrame(dfxstandard, columns=names)
```

```python
# check if data was scaled correctly
# mean and stdev should be equal to or almost equal to 0 and 1 respectively 
dfxstandard.describe()
```

```python
# notice different scale on the x-axis
# data has bee scaled correctly 
plt.hist(dfxstandard['C1'])
plt.show()
```

```python
# examine distribution for all variables 
for col in dfxstandard.columns:
    dfxstandard.hist(color='blue') #, range=(-2,2))
    break
plt.show()
```

```python
plt.boxplot(dfxstandard['C1'])
plt.xlabel('C1')
plt.show()
```

```python
# export standardized data to a csv 
dfxstandard.to_csv('csv_files/zscore.csv')
```
