```python
# import packages 
import numpy as np
import pandas as pd
```

```python
# read in file 
df = pd.read_csv('../csv_files/p1_binning.csv', index_col=0)
df.head()
```

```python
# verify that columns were renamed 
df.columns
```

```python
# filter out the x variables 
dfx = df.filter(['C1', 'C4', 'C6p', 'C7','C5p', 'T4p', 'T3p', 'T5p', 'S1p', 'S2p', 'S3p', 'C2', 'C3p', 'Industry_Bins'])
dfx.head()
```

```python
# filter out the y variables 
dfy = df.filter(['Y1', 'Y2'])
dfy.head()
```

```python
# concat dataframes 
df = pd.concat([dfx, dfy], axis=1)
df.head()
```

```python
# read dataframe to a csv file 
df.to_csv('../csv_files/p1feature-selection.csv')
```
