```python
#Importing the packages that we need 
import pandas as pd
import numpy as np
```

```python
#Loading in the csv file
data = pd.read_csv('csv_files/no_outliers.csv', header=0)
data.head()
```

```python
#To fill the NaN values in 'I3' with 0
data['I3'].fillna(0, inplace=True)
```

```python
#Filling in ID #s that had originally had multiple values 
##Found the correct ID #s on the internet
data.iloc[118, data.columns.get_loc('I3')] = 5261
data.iloc[174, data.columns.get_loc('I3')] = 7812
data.iloc[181, data.columns.get_loc('I3')] = 191
data.iloc[187, data.columns.get_loc('I3')] = 5731
data.iloc[221, data.columns.get_loc('I3')] = 7389
data.iloc[246, data.columns.get_loc('I3')] = 7372
data.iloc[499, data.columns.get_loc('I3')] = 3663
data.iloc[620, data.columns.get_loc('I3')] = 5065
```

```python
#Going to bin column 'I3'. Based on the Major Industries and then all 'others'

def assign_industry_bins(I3):
    if int(I3) >= 2000 and int(I3) <= 3999:  #Returning '1' if is in Manufacturing industry
        return(1)
    if int(I3) >= 7000 and int(I3) <= 8999:  #Returing '2' if is in Services industy
        return(2)
    else:                          #Returning '3' for all industries other than
        return(3)                  ##Manufacturing and Services

#Creating a Industry Bins column in the DataFrame
data['Industry_Bins'] = data['I3'].apply(assign_industry_bins)
```
