```python
#Importing the packages that we need 
import pandas as pd
import numpy as np
```

```python
#Loading in the csv file
dfx = pd.read_csv('csv_files/p1no_skew.csv', header=0, index_col=0)
dfx.head()
```

```python
#Loaded in the csv as a DataFrame
dfy = pd.read_csv('csv_files/y.csv', index_col=0)
dfy.head()
```

```python
#Loaded in the csv as a DataFrame
dfid = pd.read_csv('csv_files/id.csv', index_col=0)
dfid.head()
```

```python
#Concated the dfid and dfy DataFrames 
df = pd.concat([dfid, dfy], axis=1)
df.head()
```

```python
#Concated the df and dfx DataFrames
df = pd.concat([df, dfx], axis=1)
df.head()
```

```python
#To fill the '-' values in 'I3' with 0
df['I3'].replace(to_replace='-', value=0, inplace=True)
df.head(10)
```

```python
#Filling in ID #s that had originally had multiple values 
##Found the correct ID #s on the internet
df.iloc[118, df.columns.get_loc('I3')] = 5261
df.iloc[174, df.columns.get_loc('I3')] = 7812
df.iloc[181, df.columns.get_loc('I3')] = 191
df.iloc[187, df.columns.get_loc('I3')] = 5731
df.iloc[221, df.columns.get_loc('I3')] = 7389
df.iloc[246, df.columns.get_loc('I3')] = 7372
df.iloc[499, df.columns.get_loc('I3')] = 3663
df.iloc[620, df.columns.get_loc('I3')] = 5065
```

```python
#Going to bin column 'I3'. Based on the Major Industries and then all 'others'

def assign_industry_bins(I3):
    if int(I3) >= 2000 and int(I3) <= 3999:  #Returning '1' if is in Manufacturing industry
        return(1)
    if int(I3) >= 7000 and int(I3) <= 8999:  #Returing '2' if is in Services industy
        return(2)
    else:                          #Returning '3' for all industries other than
        return(3)                  ##Manufacturing and Services

#Creating a Industry Bins column in the DataFrame
df['Industry_Bins'] = df['I3'].apply(assign_industry_bins)
```

```python
#Printed out to make sure that the Binning worked 
df.head()
```

```python
#Saved to a csv
df.to_csv('csv_files/Binning_I3.csv')
```
