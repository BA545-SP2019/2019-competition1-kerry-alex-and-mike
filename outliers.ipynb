```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_clean = pd.read_csv('csv_files/clean_data_kerry.csv', index_col=0)
df_clean.head()
```

```python
df_values = df_clean.filter(items=['pIPO', 'pH', 'pL', 'p1day', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'T1', 'T2', 'T3', 'T4', 'T5', 'S1', 'S2', 'S3'])
df_values.head()                            
```

```python
# finding IQR for each variable 
Q1 = df_values.quantile(0.25)
Q3 = df_values.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
print('Q1 Values')
print(Q1)
print('Q3 Values')
print(Q3)
```

```python
# identifying outlier value locations sort of 
# Q1 - 1.5 * IQR = lower bound 
# Q3 + 1.5 * IQR = upper bound 
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR
outlier = (df_values < lower) | (df_values > upper)
# above is equivalent to this: outlier = (df_values < (Q1 - 1.5 * IQR)) | (df_values > (Q3 + 1.5 * IQR))
print(outlier)
```
