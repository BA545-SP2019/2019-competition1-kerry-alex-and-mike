```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_norm = pd.read_csv('alex_normal_data.csv', index_col=0)
df_norm.head()
```

```python
df_norm.describe()
```

```python
plt.hist(df_norm['P(IPO)'], range=(0,1))
plt.show()
```

```python
from sklearn import preprocessing
# define column names
names = df_norm.columns

# create scaler variable 
scaler = preprocessing.StandardScaler()

# fit and scale data 
df_standard = scaler.fit_transform(df_norm)
df_standard = pd.DataFrame(df_standard, columns=names)
```

```python
df_standard.describe()
```

```python
plt.hist(df_standard['P(IPO)'], range=(-2,2))
plt.show()
```

```python
df_standard.to_csv('standard_data.csv')
```
